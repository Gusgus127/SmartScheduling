```python
def clean_df(df):
    df.columns = (
        df.columns.str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('à', 'a')
        .str.replace('è', 'e')
        .str.replace('é', 'e')
        .str.replace('ò', 'o')
        .str.replace('ç', 'c')
        .str.replace('ñ', 'n')
        .str.replace('á', 'a')
        .str.replace('í', 'i')
        .str.replace('ó', 'o')
        .str.replace('ú', 'u')
    )
    for col in df.select_dtypes(include='object'):
        df[col] = df[col].str.strip()
    return df

# Load files
notes = pd.read_csv('notes_feie.csv')
access = pd.read_csv('exitacces_feie.csv')
abandon = pd.read_csv('abandonament_feie.csv')
attendance = pd.read_csv('attendance_feie.csv')

# Clean all
notes = clean_df(notes)
access = clean_df(access)
abandon = clean_df(abandon)
attendance = clean_df(attendance)

# Rename
notes = notes.rename(columns={'id_asignatura': 'codi_assignatura'})


print("Notes columns:", notes.columns.tolist())
print("Access columns:", access.columns.tolist())
print("Abandonment columns:", abandon.columns.tolist())
print("Attendance columns:", attendance.columns.tolist())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

attendance_columns = [f's{i}{j}' for i in range(1, 5) for j in range(1, 6)]
grouped = attendance.groupby(['assignatura', 'titulacio'])

cols = 5
num_plots = len(grouped)
rows = int(np.ceil(num_plots / cols))

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 4), sharey=True)
axes = axes.flatten()

for i, ((assignatura, titulacio), group) in enumerate(grouped):
    ax = axes[i]

    # Mean attendance per session (keeps session order)
    mean_attendance = group[attendance_columns].mean()

    # Interpolate missing values (only for plotting)
    interpolated = mean_attendance.interpolate(method='linear', limit_direction='both')

    # Plot: interpolated line + original points
    ax.plot(attendance_columns, interpolated, linestyle='-', linewidth=2, label='Interpolated')
    ax.scatter(attendance_columns, mean_attendance, color='black', s=10, zorder=3, label='Actual')

    ax.set_title(f'{assignatura}\n({titulacio})', fontsize=9)
    ax.set_xticks(range(len(attendance_columns)))
    ax.set_xticklabels(attendance_columns, rotation=45, ha='right', fontsize=7)

    if i % cols == 0:
        ax.set_ylabel('Avg Attendance')

    ax.grid(True)

# Hide unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# ------------------------
# Step 1: Clean and summarize attendance data
# ------------------------

# Define session columns (s11 to s45)
session_cols = [f's{i}{j}' for i in range(1, 5) for j in range(1, 6)]

# Replace NaN in session columns with 0
attendance[session_cols] = attendance[session_cols].fillna(0)

# Calculate mean attendance for each student (sum of sessions / total possible sessions)
attendance['mean_attendance'] = attendance[session_cols].sum(axis=1) / attendance['num_sessions']

# Group by subject and degree metadata, and compute average attendance and average number of sessions
attendance_summary = attendance.groupby(
    ['assignatura', 'titulacio', 'curs', 'tipologia', 'horari']
).agg({
    'mean_attendance': 'mean',
    'num_sessions': 'mean'
}).reset_index()

display(attendance_summary)

# ------------------------
# Step 2: Clean and summarize access data
# ------------------------

# Replace NaNs in access-related columns with 0 (temporary for cleaning)
access[['via_acces_estudi', "nota_d'acces_(preinscripcio)", "dedicacio_de_l'estudiant", 's/n_discapacitat', 'taxa_exit']] = \
    access[['via_acces_estudi', "nota_d'acces_(preinscripcio)", "dedicacio_de_l'estudiant", 's/n_discapacitat', 'taxa_exit']].fillna(0)

# Ensure categorical columns have clean string values or None (for mode calculation)
access['via_acces_estudi'] = access['via_acces_estudi'].apply(lambda x: x if isinstance(x, str) and x != 'nan' else None)
access["dedicacio_de_l'estudiant"] = access["dedicacio_de_l'estudiant"].apply(lambda x: x if isinstance(x, str) and x != 'nan' else None)
access['s/n_discapacitat'] = access['s/n_discapacitat'].apply(lambda x: x if isinstance(x, str) and x != 'nan' else None)

# Clean 'taxa_exit': remove '%' and convert to float for aggregation
access['taxa_exit'] = access['taxa_exit'].astype(str).str.replace('%', '', regex=False)
access['taxa_exit'] = pd.to_numeric(access['taxa_exit'], errors='coerce')

# Group access data by degree and summarize with mean or mode
access_summary = access.groupby('estudi').agg({
    'via_acces_estudi': lambda x: x.mode()[0] if not x.mode().empty else None,
    "nota_d'acces_(preinscripcio)": 'mean',
    "dedicacio_de_l'estudiant": lambda x: x.mode()[0] if not x.mode().empty else None,
    's/n_discapacitat': lambda x: x.mode()[0] if not x.mode().empty else None,
    'taxa_exit': 'mean'
}).reset_index().rename(columns={'estudi': 'titulació'})

# Reformat 'taxa_exit' back into percentage string for readability
access_summary['taxa_exit'] = access_summary['taxa_exit'].round(2).astype(str) + '%'

display(access_summary)

# ------------------------
# Step 3: Clean and summarize abandonment data
# ------------------------

# Aggregate all abandonment types by degree
abandon_summary = abandon.groupby('estudi').agg({
    'nombre_abandonaments_universitat_(1)': 'sum',
    'nombre_abandonaments_universitat_reals_(2)': 'sum',
    'nombre_abandonaments_universitat_reals_1r_any_criteri_1_any_(3)': 'sum',
    'nombre_abandonaments_universitat_reals_criteri_1_any_(4)': 'sum'
}).reset_index().rename(columns={'estudi': 'titulació'})

# Display the summary
display(abandon_summary)


```

```python
# ------------------------------------------------------------------
# 0.  Normalise `assignatura` in BOTH data sets (once and for all)
# ------------------------------------------------------------------
attendance['assignatura']      = attendance['assignatura'].str.lower()
attendance_summary['assignatura'] = attendance_summary['assignatura'].str.lower()

# ------------------------------------------------------------------
# 1.  Build a clean enrolment table (attendance_matricules)
# ------------------------------------------------------------------
attendance_feie = attendance[['assignatura',
                              'titulacio',
                              'matricules-assignatura',
                              'horari']].copy()

# Map short codes to full degree names
titulacio_mapping = {
    'ADE': "Graduat en Administració i Direcció d'Empreses",
    'ADE anglés': "Graduat en Administració i Direcció d'Empreses",
    'ADE+Dret': "Graduat en Administració i Direcció d'Empreses i Graduat en Dret",
    'EIT': 'Graduat en Empresa i Tecnologia',
    'CIF': 'Graduat en Comptabilitat i Finances',
    'ECO': 'Graduat en Economia',
    'ECO anglés': 'Graduat en Economia'
}
attendance_feie['titulacio_nom_complet'] = attendance_feie['titulacio'].map(titulacio_mapping)

# Drop exact duplicates first (same subject, degree, schedule, seats)
attendance_feie = attendance_feie.drop_duplicates(
    subset=['assignatura', 'titulacio_nom_complet', 'horari', 'matricules-assignatura']
)

# Aggregate in case the *same* key appears with split seat counts
attendance_matricules = (attendance_feie
    .groupby(['assignatura', 'titulacio_nom_complet', 'horari'], as_index=False)
    .agg({'matricules-assignatura': 'sum'})
)

# ------------------------------------------------------------------
# 2.  Merge  — one time, no leftover columns
# ------------------------------------------------------------------
attendance_summary = (attendance_summary
    .drop(columns=['matricules-assignatura'], errors='ignore')
    .merge(attendance_matricules,
           on=['assignatura', 'titulacio_nom_complet', 'horari'],
           how='left')
)

# ------------------------------------------------------------------
# 3.  Re‑compute attendance_rate safely
# ------------------------------------------------------------------
attendance_summary['attendance_rate'] = (
    attendance_summary['mean_attendance'] /
    attendance_summary['matricules-assignatura']
)

# Optional sanity check
print("Rows:", len(attendance_summary),
      "|  missing matricules:", attendance_summary['matricules-assignatura'].isna().sum())

display(attendance_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.colors import BoundaryNorm
from matplotlib.cm import get_cmap

attendance_summary['horari'] = pd.Categorical(
    attendance_summary['horari'],
    categories=sorted(attendance_summary['horari'].unique()),
    ordered=True
)

horari_positions = attendance_summary['horari'].cat.codes

boundaries = np.arange(0, 1.01, 0.2)  # from 0 to 1 in steps of 0.2
cmap = get_cmap('Blues', len(boundaries) - 1)
norm = BoundaryNorm(boundaries, ncolors=cmap.N, clip=True)

plt.figure(figsize=(4, 6))
scatter = plt.scatter(
    horari_positions,
    attendance_summary['nota_mitjana_asignatura'],
    c=attendance_summary['attendance_rate'],
    cmap=cmap,
    norm=norm,
    edgecolors='w',
    s=100,
    alpha=0.8
)

plt.ylim(0, 10)
plt.xticks(
    ticks=range(len(attendance_summary['horari'].cat.categories)),
    labels=attendance_summary['horari'].cat.categories,
    ha='center'
)

cbar = plt.colorbar(scatter, boundaries=boundaries, ticks=boundaries)
cbar.set_label('Attendance Rate')

plt.xlabel('Horari')
plt.ylabel('Nota Mitjana Assignatura')
plt.title('Marks vs Horari')
plt.tight_layout()
plt.show()
```
